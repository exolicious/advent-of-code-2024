```python
def parse_input(input):
    grid_string = input
    grid = np.array([list(line) for line in grid_string.splitlines()])
    
    return grid
```

```python
def rotate_clockwise(direction):
    mapping = {
        "up": "right",
        "right": "down",
        "down": "left",
        "left": "up"
    }
    return mapping[direction]

def rotate_counterclockwise(direction):
    mapping = {
        "up": "left",
        "left": "down",
        "down": "right",
        "right": "up"
    }
    return mapping[direction]

direction_vectors = {
    "up": np.array([-1, 0]),
    "down": np.array([1, 0]),
    "left": np.array([0, -1]),
    "right": np.array([0, 1])
}

def get_cheapest_path_cost(paths):
    if not paths:
        return float('inf')
    lowest_value = paths[0][1]
    for element in paths:
        if element[1] < lowest_value:
            lowest_value = element[1]
    return lowest_value
```

```python
import heapq

grid = parse_input(input)

def get_unique_node_positions(all_legit_paths):
    unique_positions = set()
    for state in all_legit_paths:
        unique_positions.add(state[0])
    print(len(unique_positions))


def get_cheapest_path(grid):
    lowest_cost = float('inf')
    start_pos = (np.where(grid == "S")[0][0], np.where(grid == "S")[1][0])
    all_nodes_of_lowest_cost_paths = set()
    best_costs = {}
    pq = [(0, start_pos, "right", [])]
    
    while pq:
        cost, pos, direction, path = heapq.heappop(pq)
        
        state = (pos, direction)

        if cost > lowest_cost: # global cost check
            continue

        if grid[pos[0], pos[1]] == "E" and cost == lowest_cost:
            all_nodes_of_lowest_cost_paths.update(path + [pos])
        elif grid[pos[0], pos[1]] == "E" and cost < lowest_cost:
            lowest_cost = cost
            all_nodes_of_lowest_cost_paths = set(path + [pos])

        if state in best_costs and cost > best_costs[state]: # node specific (pos and direction) cost check, if there is a path that has a better cost already for this pos and direction combination, continue
            continue

        best_costs[state] = cost

        next_node_in_initial_direction_pos = (pos[0] + direction_vectors[direction][0], pos[1] + direction_vectors[direction][1])
        clockwise_rotated_direction = rotate_clockwise(direction)
        clockwise_rotated_next_node_pos = (pos[0] + direction_vectors[clockwise_rotated_direction][0], pos[1] + direction_vectors[clockwise_rotated_direction][1])
        counterclockwise_rotated_direction = rotate_counterclockwise(direction)
        counterclockwise_rotated_next_node_pos = (pos[0] + direction_vectors[counterclockwise_rotated_direction][0], pos[1] + direction_vectors[counterclockwise_rotated_direction][1])

        if grid[next_node_in_initial_direction_pos[0], next_node_in_initial_direction_pos[1]] != "#":
            local_path = path.copy()
            local_path.append(next_node_in_initial_direction_pos)
            heapq.heappush(pq, (cost + 1, next_node_in_initial_direction_pos, direction, local_path))
        if grid[clockwise_rotated_next_node_pos[0], clockwise_rotated_next_node_pos[1]] != "#":
            local_path = path.copy()
            local_path.append(clockwise_rotated_next_node_pos)
            heapq.heappush(pq, (cost + 1001, clockwise_rotated_next_node_pos, clockwise_rotated_direction, local_path))
        if grid[counterclockwise_rotated_next_node_pos[0], counterclockwise_rotated_next_node_pos[1]] != "#":
            local_path = path.copy()
            local_path.append(counterclockwise_rotated_next_node_pos)
            heapq.heappush(pq, (cost + 1001, counterclockwise_rotated_next_node_pos, counterclockwise_rotated_direction, local_path))
    print(all_nodes_of_lowest_cost_paths)
    return len(all_nodes_of_lowest_cost_paths)

print(get_cheapest_path(grid))

```
