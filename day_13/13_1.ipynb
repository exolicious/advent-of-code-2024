```python
def parse_input_to_array(input_text):
    lines = input_text.strip().split("\n")
    results = []
    A = []
    b = []

    for line in lines:
        if line.startswith("Button A"):
            parts = line.split(", ")
            a_x = int(parts[0].split("+")[1])
            a_y = int(parts[1].split("+")[1])
        elif line.startswith("Button B"):
            parts = line.split(", ")
            b_x = int(parts[0].split("+")[1])
            b_y = int(parts[1].split("+")[1])
            A = [[a_x, b_x], [a_y, b_y]]
        elif line.startswith("Prize"):
            parts = line.split(", ")
            x_prize = int(parts[0].split("=")[1])
            y_prize = int(parts[1].split("=")[1])
            b = [x_prize, y_prize]
            # Append the current block as a tuple
            results.append((np.array(A), np.array(b)))

    return results

input_list = parse_input_to_array(input)
print(input_list)
```
